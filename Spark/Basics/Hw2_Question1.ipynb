```
from io import StringIO
```

```
inp = sc.textFile("/FileStore/tables/mutual.txt")
```

```
user_friends = inp.map(lambda line: line.split("\t")) # creates list of lists with each list being a user and then their friends
user_friends.take(5)
```

```
split_friends = user_friends.map(lambda arr: [arr[0], arr[1].split(",")]) # separates friends in each list inside list of lists 

split_friends.take(5)
# split_friends.coalesce(1).saveAsTextFile("/FileStore/tables/split_friends.txt")
```

```
def create_pairs(line):
    user = line[0]
    friends = line[1]    
    
    pair_friends = []

    for f in friends:
        if friends == ['']:
            continue
        if int(user) < int(f):
            pair_friends.append(((user, f), friends))
        else:
            pair_friends.append(((f, user), friends))
            
    return pair_friends
    
```

```
pair_friends = split_friends.flatMap(create_pairs)
pair_friends.take(5)
# pair_friends.coalesce(1).saveAsTextFile("FileStore/tables/pairs_friends1.txt")
```

```
def mutual_friends(f1, f2):
    mutuals = set(f1)
    mutuals = mutuals.intersection(set(f2))

    return len(mutuals)
```

```
mutuals = pair_friends.reduceByKey(mutual_friends)
mutuals.take(20)

# dbutils.fs.rm("dbfs:/FileStore/tables/mutual_friends.txt", True)
# mutuals.coalesce(1).saveAsTextFile("FileStore/tables/mutual_friends.txt")
```

```
# with open('/tmp/mutual_friends.txt', 'w') as f:
for line in mutuals.take(5):
    output = f"{line[0][0]}, {line[0][1]}\t{line[1]}"
    print(output)
#         f.write(output)
    
```

```
# outputting to file

with StringIO() as file:
        
    for line in mutuals.collect():
        
        user1 = line[0][0]
        user2 = line[0][1]
        mutual_number_of_friends = line[1]
        
        output = f"{user1}, {user2}\t{mutual_number_of_friends}\n"
        
        file.write(output)
        
    dbutils.fs.put("/FileStore/tables/Hw2_Q1_output.txt", file.getvalue(), True)
```
